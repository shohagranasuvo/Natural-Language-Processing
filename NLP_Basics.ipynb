```python
import spacy
```

```python
nlp =spacy.load("en_core_web_sm")
doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulkk loves chat of delhi .")
for sentence in doc.sents:
    print(sentence)
    for word in sentence:
        print(word)
        
```
