```python
import spacy 
import nltk
```

```python
from nltk import PorterStemmer
stemmer =PorterStemmer()
```

```python
words =["eating","Ability","geting","offcourse","iies"]
```

```python
for word in words:
    print (word ," | " ,stemmer.stem(word))
```

```python
nlp =spacy.load("en_core_web_sm")
```

```python
doc= nlp("eating eats eat ate adjustable rafting ability meeting better is")
for  word in doc :
    print(word ," | " ,word.lemma_  ," | " ,word.lemma) 
```

```python
nlp.pipe_names
```

```python
ar  =nlp.get_pipe('attribute_ruler')
ar.add([[{"TEXT" :"bro"}],[{"TEXT":"brah"}]],{"LEMMA":"Brother"})

doc=nlp ("bro what's its going there ?? brah just wanna to know ? ")
for token in doc :
     print (token.text , " | " ,token.lemma_)
    
    
```
