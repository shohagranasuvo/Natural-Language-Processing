```python
nlp = spacy.load("en_core_web_trf")
```

```python
def tokenize(text):
    if pd.isna(text):
        return []
    doc = nlp(str(text))
    return [token.text for token in doc]
df["tokens"] =df["Text"].apply(tokenize)
```
